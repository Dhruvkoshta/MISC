```python
# prompt: laplace transform of sint from 0 to pi/2

from sympy import *
t,s,a,b = symbols('t,s,a,b')
y = sin(t)
Y = laplace_transform(y,t,s, noconds=True)
display(simplify(Y))
lower_limit = 0
upper_limit = pi/2
result = integrate(y * exp(-s*t), (t, lower_limit, upper_limit))
display(result)
```
